```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do sistema
N = 51  # número de células (ímpar para ter um centro definido)
steps = 2 * N  # número de passos de evolução

# Estado inicial: tudo -1 exceto célula central com +1
x = -np.ones(N, dtype=int)
x[N // 2] = 1

# Armazena a evolução do sistema
history = np.zeros((steps, N), dtype=int)
history[0] = x.copy()

# Função para evoluir uma geração do autômato
def evolve(x):
    x_new = np.empty_like(x)
    for i in range(1, N - 1):
        x_new[i] = x[i - 1] * x[i + 1] * (-1)
    x_new[0] = x[0]  # bordas fixas (sem evolução)
    x_new[-1] = x[-1]
    return x_new

# Evolução do sistema
for t in range(1, steps):
    x = evolve(x)
    history[t] = x.copy()

# Visualização da evolução
plt.figure(figsize=(12, 6))
plt.imshow(history.T, cmap="Greys", interpolation="nearest", aspect="auto")
plt.xlabel("Tempo")
plt.ylabel("Índice da célula")
plt.title("Evolução de um autômato celular 1D")
plt.colorbar(label="Estado da célula (+1 = preto, -1 = branco)", ticks=[-1, 1])
plt.show()
```
